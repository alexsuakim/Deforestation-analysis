```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#annual deforestation trend
defo_df = pd.read_csv(r"./deforestation\def_area_2004_2019.csv")
states = ['Acre', 'Amazonas', 'Amapa', 'Maranhao', 'Mato Grosso', 'Para', 'Rondonia', 'Roraima', 'Tocantins']
fig, axes = plt.subplots(figsize=(15,8))

defo_df.plot(x='Ano/Estados', y=defo_df.columns[1:10], 
                 label=states , colormap='tab10', linewidth=2, ax=axes)

axes.set_xlabel('Year', fontsize=12)
axes.set_ylabel('Area in $km^2$', fontsize=12)
axes.set_title('Annual Area of Deforested Amazon Rainforest by State 2004-2019', fontsize=14)
axes.grid(axis='y', color='0.95')
```

```python
#read deforestation area
defo_df = pd.read_csv(r"./deforestation\def_area_2004_2019.csv")
defo_df.drop(['AMZ LEGAL'], axis=1, inplace=True)

defo_df = pd.melt(defo_df, id_vars=["Ano/Estados"], 
        var_name="state", 
        value_name="value")
defo_df.columns = ['year', 'state', 'area']

#read precipitation
df_precipitation = pd.read_csv(r"./precipitation\precipitation.csv")
df_precipitation['date'] = df_precipitation['date'].astype(str).str[6:10]
df_precipitation['date'] = df_precipitation['date'].astype(int)
df_precipitation = df_precipitation[df_precipitation['date'] >= 2004]

prec_df = df_precipitation.groupby(['date','state'])['precipitation'].mean().to_frame(name = 'precipitation').reset_index()
prec_df.columns = ['year', 'state', 'precipitation']

#inner join deforestation area and precipitation
joined_df = pd.merge(defo_df, prec_df)
joined_df
```

```python
#divide df by states
acre = joined_df[joined_df['state'] == 'AC']
amazonas = joined_df[joined_df['state'] == 'AM']
amapa = joined_df[joined_df['state'] == 'AP']
maranhao = joined_df[joined_df['state'] == 'MA']
mato_grosso = joined_df[joined_df['state'] == 'MT']
para = joined_df[joined_df['state'] == 'PA']
roraima = joined_df[joined_df['state'] == 'RR']
tocantins = joined_df[joined_df['state'] == 'TO']
```

```python
#plot 
fig, ax = plt.subplots(figsize = (10, 5))
plt.title('Acre')
x = acre['year']
y1 = acre['area']
y2 = acre['precipitation']
ax2 = ax.twinx()
ax.plot(x, y1, color = 'g')
ax2.plot(x, y2, color = 'b')
ax.set_xlabel('year')
ax.set_ylabel('deforested area', color = 'g')
ax2.set_ylabel('precipitation', color = 'b')
plt.tight_layout()
plt.show()

#correlation between area and precipitation in Acre.
print(acre['area'].corr(acre['precipitation']))
```
